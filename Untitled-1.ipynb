```python
from src.database import Produto, SessionLocal

def registrar_entrada_produto(db_session, codigo, nome, quantidade, valor=0.0, data=None, fornecedor=None, local=None):
    """
    Registra a entrada de um produto no estoque (Lógica Pura).
    """
    if not nome or quantidade <= 0:
        raise ValueError("Dados inválidos: Nome obrigatório e quantidade deve ser positiva.")

    produto = db_session.query(Produto).filter_by(codigo=codigo).first()

    if produto:
        produto.quantidade += quantidade
        is_new = False
    else:
        produto = Produto(
            codigo=codigo,
            nome=nome,
            quantidade=quantidade,
            data=data,
            fornecedor=fornecedor,
            local=local,
            valor=valor
        )
        db_session.add(produto)
        is_new = True
    
    db_session.commit()
    return produto, is_new
```
